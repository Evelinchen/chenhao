```python
'''

1、去除空值
2、统计每个种类的频数


'''
people = gl_data['people'][gl_data['people']!='None']
cost = gl_data['cost'][gl_data['cost']!='None'].apply(lambda x: int(x.split('/')[1][0:-3]))
days = gl_data['day'][gl_data['day']!='None'].apply(lambda x: int(x.split('/')[1][0:-2]))
day_count = {}
for i in days:
    if i in day_count:
        day_count[i]+=1
    else:
        day_count[i] = 1
# day_count
people_count = {}
for i in people:
    if i in people_count:
        people_count[i] += 1
    else:
        people_count[i] = 1
# people_count
cost_count = {}
for i in cost:
    if i in cost_count:
        cost_count[i] += 1
    else:
        cost_count[i] = 1
# day_count
```

## 三、和谁一起去

#### 第一天：成都—康定
早晨出发，经成雅高速、二郎山隧道、泸定县（电站施工，实施交通管制，对交通影响较大），所以到康定就不走了，简单的感受下情歌故乡，寻找溜溜大姐和帅气的康巴汉子。
宿：康定
#### 第二天：康定—新都桥—塔公—雅江
康定出发就翻越关外名山—折多山（海拔4298米），稍做停留，拍照纪念就下山，新都桥周围的风光无限。
新都桥，处在川藏线南北分叉路口，是一片如诗如画的世外桃源。神奇的光线，无垠的草原，弯弯的小溪，金黄的柏杨，山峦连绵起伏，藏寨散落其间，牛羊安详地吃草……川西的平原风光美丽地绽放。这，就是新都桥，令人神往的“光与影的世界”、“摄影的天堂”。

早起拍摄神山—央迈勇，晨曦下的神山央迈勇慑人心魄，找片水作前景拍倒影，水里的央迈勇更美（注意躲开管理员的监视）；徜徉在三座神山和美丽的牛场共同组成的景致中（想象一下吧）；天气及体力许可的情况下，经贡嘎殿前往牛奶海、五色海；（这天中午的路餐个人自行准备）之后返回牛场；休整之后来到冲古寺再次拍摄日落的夏诺多吉神山；之后回到龙龙坝乘车返回稻城再去泡泡有名的茹布查卡温泉，洗去一身希脏，还有满身疲惫；

<font color=#0099ff size=6 face="黑体"> 六、游记文本分析 </font>

```python
##1先来分词吧
import jieba
# type(''.join(gl_data['note_text'][0:3]))
# f.write(''.join(gl_data['note_text'][gl_data['note_text']!='None'].apply(lambda x: x[1:-1])),enco)
'''
1、把数据框里面note_text(游记)整合成一篇长文章
2、对长文章切词
3、统计词频
'''
all_text = ''.join(gl_data['note_text'][gl_data['note_text']!='None'].apply(lambda x: x[1:-1]))    ##1
cut_text = jieba.cut(all_text)                                                                     ##2
word_dic = {}                                                                                      ##3
for i in cut_text:
    if i in word_dic:
        word_dic[i] +=1
    else:
        word_dic[i] = 1
```

```python
'''
应为切分好的词语里面存在大量，语气用词、标点符号、不规则符号，我们需要把这些清除掉
'''
file = open('stop_word.txt')
st_w = file.read().split()
cleaned = {}
for i in word_dic:
    if i not in st_w:
        cleaned[i] = word_dic[i]
```

```python
pd.Series(cleaned).to_csv('cleand_word_dict.csv',encoding="utf_8_sig")
```
